### Working Example 

```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
import string

#####__________train dataset__________#####
df_train = pd.DataFrame(np.arange(120).reshape(40,3),
                  columns=list('ABC')).applymap(lambda x: np.random.choice(list(string.ascii_letters)))

#####__________inference dataset__________#####
df_infer = pd.DataFrame(np.arange(12).reshape(4,3),
                  columns=list('ABC')).applymap(lambda x: np.random.choice(list(string.ascii_letters)))
```

```python
##### __________List all columns that needs to be label encoded__________#####
categorical_features =['A', 'B', 'C']

##### __________label encode the train dataset__________#####
df_train_lbl, encoder_dict = Label_Encoder_fit(df_train, categorical_features)
```

```python
#####__________Pickle the encoder dict__________#####
joblib.dump(encoder_dict, 'encoder.pkl')
```

```python
##### __________label encode the Inference dataset using the encoder pickle__________#####
df_infer_lbl = label_encode_transform(df_infer = df_infer.copy(), 
                                      encoder='C:\\Users\\Timoth\\Desktop\\AMN\\encoder.pkl',
                                      categorical_features = categorical_features)
```
